```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
columns_name = ["user_id", "item_id", "rating", "timestamp"]
```

```python
df = pd.read_csv("u.data", sep="\t", names=columns_name)
df.head()
```

```python
movie_titles = pd.read_csv("Movie_Id_Titles")
movie_titles.head(3)
```

```python
df = pd.merge(df,movie_titles,on="item_id")
df.head()
```

```python
df.groupby(["title"])["rating"].mean().sort_values(ascending = False).head()
```

```python
df.groupby("title")["rating"].count().sort_values(ascending = False).head(10)
```

```python
ratings = pd.DataFrame(df.groupby(["title"])["rating"].mean())
ratings.head()
```

```python
ratings["num of ratings"] = pd.DataFrame(df.groupby("title")["rating"].count())
ratings.head()
```

```python
ratings["num of ratings"].hist(bins = 50)
```

```python
ratings["rating"].hist(bins = 50)
```

```python
sns.scatterplot(x= "rating", y = "num of ratings", data= ratings)
```

```python
movie_matrix = df.pivot_table(index = "user_id", columns="title", values = "rating")
movie_matrix.head()
```

```python
ratings.sort_values("num of ratings", ascending = False).head()
```

```python
star_war_rating = movie_matrix["Star Wars (1977)"]
liar_liar_rating = movie_matrix["Liar Liar (1997)"]
```

```python
star_war_rating.head()
```

```python
similar_to_starwar =  movie_matrix.corrwith(star_war_rating)
similar_to_liarliar = movie_matrix.corrwith(liar_liar_rating)
```

```python
corr_starwar = pd.DataFrame(similar_to_starwar, columns=["Correlation"])
corr_starwar.dropna(inplace = True)
corr_starwar.head()
```

```python
corr_starwar = corr_starwar.join(ratings["num of ratings"])
corr_starwar.head()
```

```python
corr_starwar[corr_starwar["num of ratings"] > 100].sort_values("Correlation", ascending = False)
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=["Correlation"])
corr_liarliar.dropna(inplace = True)
corr_liarliar = corr_liarliar.join(ratings["num of ratings"])
corr_liarliar.head()
```

```python
corr_liarliar[corr_liarliar["num of ratings"] > 100].sort_values("Correlation", ascending = False)
```
